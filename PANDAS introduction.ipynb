# seaborn

```python
import matplotlib.pyplot as plt
import  seaborn as sns
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = 9,5
import warnings
warnings.filterwarnings('ignore')
```

```python
stats.head()
```

```python
vis1 = sns.distplot(stats["InternetUsers"]) # univariant 
plt.show()
```

```python
vis1 = sns.distplot(stats["InternetUsers"], bins = 10)
plt.show()
```

```python
vis2 = sns.boxplot( data = stats, x="IncomeGroup", y='BirthRate')
plt.show()
```

```python
vis4 = sns.lmplot(data = stats, x = 'InternetUsers', y = 'BirthRate')
plt.show()
```

```python
vis3 = sns.lmplot(data = stats, x = 'InternetUsers', y ='BirthRate', fit_reg = True)
plt.show()
```

```python
vis31 = sns.lmplot(data = stats, x = 'InternetUsers', y = 'BirthRate', fit_reg =  False) # lm - linear model.
plt.show()
```

```python
vis5 = sns.lmplot(data = stats, x = 'InternetUsers', y = 'BirthRate',fit_reg = False , hue = 'IncomeGroup')
plt.show()
```

```python
vis5 =   sns.lmplot(data = stats, x = 'InternetUsers', y = 'BirthRate',fit_reg = True, hue = 'IncomeGroup', size = 7)
plt.show()
```
